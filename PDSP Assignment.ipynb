```python
def squarfree(x):
        output=True
        i=2
        while i<=x//2 and output==True:
            y=x/i
            if int(y)==y:
                m=y/i
                if int(m)==m:
                    output=False
            i+=1
        return output
            
```

```python
def perfect(x):
    sum1=1
    i=2
    while i<=x//2:
        d=x/i
        if int(d)==d:
            sum1+=i
        i+=1    
    return d
        
```

```python
def nondecreasing(l):
    x=True
    for i in range(len(l)):
        for j in range(0,i):
            if l[j]>l[i]:
                x=False
                break
    return x

                
                
        
```

```python
def matched(s):
    b=0
    for i in s:
        if i==')':
            b+=-1
            if b==-1:
                return False
                
        elif i=='(':
            b+=1
    if b==0:
        return True   
    else:
        return False
        
        
```

```python
def expanding(l):
    x=True
    min1=abs(l[1]-l[0])
    for i in range(2,len(l)):
        dis=abs(l[i]-l[i-1])
        if dis<=min1:
            return False
        min1=dis
    return x   
      
        
```

```python
def sumsquare(x):
    l=[0,0]
    for i in x:
        if i%2==0:
            l[1]+=i*i
        else:
            l[0]+=i*i
    return l
```

```python
def counthv(l):
    h=[0,0]
    for i in range(1,len(l)-1):
        if l[i+1]<l[i]>l[i-1]:
            h[0]+=1
        elif l[i+1]>l[i]<l[i-1]:
            h[1]+=1
    return h        
            
    
```

```python
def rainaverage(l):
    l.sort()
    l1=[]
    l2=[]
    c=[]
    for i,n in l:
        if i in l1:
            j=l1.index(i)
            l2[j]+=n
            c[j]+=1
        else:
            l1.append(i)
            l2.append(n)
            c.append(1)
    result=[]
    for i in range(len(l1)):
        result.append((l1[i],l2[i]/c[i]))
    
    return result
```

```python
def matrixflip(x,y):
    if y=='v':
        return x[::-1]
    elif y=='h':
        b=[]
        for i in x:
            
            b.append(i[::-1])
        return b    
    else:
        return x
```

```python
def rotate(x):
    ret=[]
    for j in range(len(x)):
        m=[]
        for i in range(1,len(x)+1):
            m.append(x[-i][j])
        ret.append(m)  
    return ret      
```

```python
def dec_inc(l):
    if len(l)<=1:
        return(True)
    else:
        return(l[1]>l[0] and inc_dec(l[1:]))
def inc_dec(l):
    if len(l)<=1:
        return(True)
    else:
        return(l[1]<l[0] and dec_inc(l[1:]))

    
def zigzag(l):
    return(inc_dec(l) or dec_inc(l))

def accordian(l):
    if len(l)<=1:
        return(True)
    else:
        diff=[]
        for i in range(len(l)-1):
            diff.append(abs(l[i]-l[i+1]))
    return(zigzag(diff))
```

```python
def downup(l):
    if len(l)<=1:
        return(True)
    else:
        return(l[1]>l[0] and updown(l[1:]))
def updown(l):
    if len(l)<=1:
        return(True)
    else:
        return(l[1]<l[0] and downup(l[1:]))

    
def zigzag(l):
    return(updown(l) or downup(l))

def accordian(l):
    if len(l)<=1:
        return(True)
    else:
        diff=[]
        for i in range(len(l)-1):
            diff.append(abs(l[i]-l[i+1]))
    return(zigzag(diff))
```

```python
def histogram(l):
    l.sort()
    dict1={}
    for i in l:
        if i in dict1.keys():
            dict1[i]+=1
        else:
            dict1[i]=1
    l1= list(sorted(dict1.items(), key=lambda item: item[1]))     
    return l1

def onehop(l):
    l=sorted(l)
    x=[]
    l1=[]
    for i in range(len(l)):
        for j in range(len(l)):
            if l[i][1]==l[j][0]  :
                if l[i][0]!=l[j][1]:
                    l1.append((l[i][0],l[j][1]))
    x=sorted(list(set(l1)))                           
    return x  
def polysum(l):
    dict1={}
    for x,y in l:
        if y in dict1.keys():
            dict1[y]+=x
        else:
            dict1[y]=x
    l=[(dict1[x],x) for x in dict1.keys() if dict1[x] != 0 ]        
    l=sorted(l,key=lambda x: x[1],reverse=True)
    return l  
def addpoly(l1,l2):
    return polysum(l1+l2)    

def multpoly(l1,l2):
    l3=[(x*p,y+q) for x,y in l1 for p,q in l2 ]
    return polysum(l3)
```
